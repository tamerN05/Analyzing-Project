```python
for col_4 in global_.columns:
    unique_4 = global_[col_4].unique()[:50]
    print(col_4,unique_4)
```

```python
temp_4 = global_.dropna(subset= 'LandAverageTemperature').copy()
temp_4 = temp_4.dropna(subset= 'LandAverageTemperatureUncertainty')
temp_4 = temp_4.dropna(subset= 'LandMaxTemperature')
temp_4 = temp_4.dropna(subset= 'LandMaxTemperatureUncertainty')
temp_4 = temp_4.dropna(subset= 'LandMinTemperature')
temp_4 = temp_4.dropna(subset= 'LandMinTemperatureUncertainty')
temp_4 = temp_4.dropna(subset= 'LandAndOceanAverageTemperature')
temp_4 = temp_4.dropna(subset= 'LandAndOceanAverageTemperatureUncertainty')

median_L = temp_4['LandAverageTemperature'].median()
median_LU = temp_4['LandAverageTemperatureUncertainty'].median()
median_LM = temp_4['LandMaxTemperature'].median()
median_LMU = temp_4['LandMaxTemperatureUncertainty'].median()
median_Lm = temp_4['LandMinTemperature'].median()
median_Lmu = temp_4['LandMinTemperatureUncertainty'].median()
median_LO = temp_4['LandAndOceanAverageTemperature'].median()
median_LOU = temp_4['LandAndOceanAverageTemperatureUncertainty'].median()
```

```python
global_['LandAverageTemperature'] = global_['LandAverageTemperature'].fillna(median_L)
global_['LandAverageTemperatureUncertainty'] = global_['LandAverageTemperatureUncertainty'].fillna(median_LU)
global_['LandMaxTemperature'] = global_['LandMaxTemperature'].fillna(median_LM)
global_['LandMaxTemperatureUncertainty'] = global_['LandMaxTemperatureUncertainty'].fillna(median_LMU)
global_['LandMinTemperature'] = global_['LandMinTemperature'].fillna(median_Lm)
global_['LandMinTemperatureUncertainty'] = global_['LandMinTemperatureUncertainty'].fillna(median_Lmu)
global_['LandAndOceanAverageTemperature'] = global_['LandAndOceanAverageTemperature'].fillna(median_LO)
global_['LandAndOceanAverageTemperatureUncertainty'] = global_['LandAndOceanAverageTemperatureUncertainty'].fillna(median_LOU)

global_
```
